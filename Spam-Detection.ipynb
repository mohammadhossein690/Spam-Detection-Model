# Spam Detection
**NLP classification project using Naive Bayes**


## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('spam.csv')
```

## Cleaning the texts

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus = []
for i in range(0,5572):
    message = re.sub('[^a-zA-Z]',' ',dataset['Message'][i])
    message = message.lower()
    message = message.split()
    ps = PorterStemmer()
    all_stopwords = stopwords.words('english')
    message = [ps.stem(word) for word in message if word not in set(all_stopwords)]
    message = ' '.join(message)
    corpus.append(message)
```

## Creating the Bag of Words model

```python
# This is called Tokenization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=6290)
X = cv.fit_transform(corpus).toarray()
y = dataset['Category'].values
```

```python
len(X[0])
```

## Encoding Dependant Variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

## Training the Naive Bayes model on the Training set

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB(alpha=2,fit_prior=True)
classifier.fit(X_train,y_train)
```

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(f'The accuracy of test set is {accuracy_score(y_test,y_pred)}')
```

## Applying k-Fold Cross Validation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=classifier,X=X_train,y=y_train,cv=10,scoring='accuracy')
print(f'The mean of 10 folds accuracies is {accuracies.mean()}')
print(f'The std of 10 folds accuracies is {accuracies.std()}')
```
